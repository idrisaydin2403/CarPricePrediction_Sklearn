## REMOVING UNNECESSARY COLUMNS

```python
df = df.drop(columns = ['car_ID', 'CarName', 'doornumber', 'carbody', 'drivewheel', 'enginelocation', 'fuelsystem', 'cylindernumber', 'enginetype'])
```

```python
df
```

## CONVERTING CATEGORICAL VARIABLES TO NUMERIC VALUES

```python
df = pd.get_dummies(df, columns = ['fueltype', 'aspiration'], drop_first=True).astype(int)
```

```python
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
corr_matrix = df.corr()
```

## VISUALIZATON OF CORRELATION MATRIX WITH HEATMAP

```python
plt.figure(figsize = (13, 13))
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm')
plt.show()
```

## CREATING TRAIN AND TEST SETS

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns = 'price')
y = df['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

## CREATING THE MODEL

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

## TRAINING THE MODEL

```python
model.fit(X_train, y_train)
```

## EVALUATING MODEL PERFORMANCE

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
prediction = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, prediction)
mae = mean_absolute_error(y_test, prediction)
r2 = r2_score(y_test, prediction)

print(f"Mean Squared Error:{mse}")
print(f"Mean Absolute Error:{mae}")
print(f"R^2 Score:{r2}")
```

```python
sns.set_theme()

plt.figure(figsize = (10, 7))
sns.scatterplot(x=y_test, y=prediction)
plt.title('\nTrue Values - Estimated Values\n')
plt.xlabel('\nTrue Values')
plt.ylabel('Estimated Values\n')

# Diyagonal doğru ekleme
min_val = min(min(y_test), min(prediction))
max_val = max(max(y_test), max(prediction))
plt.plot([min_val, max_val], [min_val, max_val], color='red', linestyle='--')

plt.show()
```
